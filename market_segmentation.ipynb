```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("D:/courses/[FreeCourseSite.com] Udemy - The Data Science Course Complete Data Science Bootcamp/38 - Advanced Statistical Methods - K-Means Clustering/011 3.12.Example.csv")
data
```

```python
plt.scatter(data["Satisfaction"], data["Loyalty"])
plt.xlabel("Satisfaction")
plt.ylabel("Loyalty")
```

```python
#selecting features:
x = data.copy()
```

```python
#clustering:
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
#clustering results:
clusters = x.copy()
clusters["cluster_prediction"] = kmeans.fit_predict(x)
```

```python
plt.scatter(clusters["Satisfaction"], clusters["Loyalty"], c = clusters["cluster_prediction"], cmap = "rainbow")
plt.xlabel("Satisfaction")
plt.ylabel("Loyalty")
```

Standardize the Variables:

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)   #converts mean to 0 and standard deviation to 1
x_scaled
```

Elbow Method:

```python
wcss = []
for i in range(1, 10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
wcss
```

```python
plt.plot(range(1, 10), wcss)
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
```

Selecting Number of Clusters:

```python
kmeans_new = KMeans(2)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new["cluster_prediction"] = kmeans_new.fit_predict(x_scaled)
clusters_new
```

```python
plt.scatter(clusters_new["Satisfaction"], clusters_new["Loyalty"], c = clusters_new["cluster_prediction"], cmap = "rainbow")
plt.xlabel("Satisfaction")
plt.ylabel("Loyalty")
```
